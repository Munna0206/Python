```python
# 1.Create a class called Employee with member data empno, name, depname, designation, age and salary and perform the following function
# 1.To accept details of 5 employees
# 2.To display details of all the employees
# 3.To search for an employee among all employees and display the details of that employee
# 4.To find the total salary of all the employee.

class Employee():
    def __init__(self, empno, ename, depname, designation, age, salary):
        self.empno = empno
        self.ename = ename
        self.depname = depname
        self.designation = designation
        self.age = age
        self.salary = salary

    def printEmp(self):
        print("\nEmployee number:", self.empno)
        print("Employee name:", self.ename)
        print("Employee department name:", self.depname)
        print("Employee designation:", self.designation)
        print("Employee age:", self.age)
        print("Employee salary:", self.salary)

emp = []
total=0
for i in range(3):
    print("\nEnter the details of employee", i+1, "\n")
    empno = int(input("Enter the employee number:"))
    ename = input("Enter the employee name:")
    depname = input("Enter the employee department:")
    designation = input("Enter the employee designation:")
    age = int(input("Enter the employee age:"))
    salary = int(input("Enter the employee salary:"))
    total=total+salary
    emp.append(Employee(empno, ename, depname, designation, age, salary))

while True:
    ch = int(input("\n1.Display all the employees\n2.search for an employee\n3.find total salary of all employee\n4.exit\nEnter your choice:"))
    if ch == 1:
        for i in range(3):
            emp[i].printEmp()
    elif ch == 2:
        name = input("Enter the employee name to search:")
        index = None
        for i in range(3):
            if name == emp[i].ename:
                index = i
                emp[index].printEmp()
                break
        if index != None:
            print("Employee not found")
    elif ch == 3:
        print("Total salary of all employee is:", total)
    elif ch == 4:
        break
    else:
        print("Invalid choice")
```

```python
# Create a class called Library with Accno, Callno, Title of the book, author, price of the book. Perform the following
# 1.	To accept the details of 5 books and store it in database.
# 2.	To display the details of all books.
# 3.	To find the book with highest cost and display the details of that book.
# 4.   To display the total cost of all books put together.
# 5.   To search for a perticular book


import sqlite3
#Database Connection
cursor=sqlite3.connect("empdata.db")
cursor.execute("create table if not exists book(accno integer not null,callno integer not null,title text not null,author text not null,price integer not null)")

class Library():
    def add(self, accno, callno, title, author, price):
        cursor.execute("insert into book values(?,?,?,?,?)",(int(accno),int(callno),title,author,int(price)))
        cursor.commit()
    def disp(self):
        row=cursor.execute("select * from book")
        for i in row:
            print(i)
    def max(self):
        row=cursor.execute("select * from book")
        max=0
        for i in row:
            if(i[4]>max):
                max=i[4]
                dct=i
        print("\nThe book with maximum cost is:\n",dct)
    def total(self):
        row=cursor.execute("select * from book")
        total=0
        for i in row:
            total+=i[4]
        print(total)
    def srch(self,book):
        row=cursor.execute("select * from book where title="+str(book))
        for i in row:
            print(i)


lib=Library()

while True:
    ch = int(input("\n1.Accept Details\n2.Display All\n3.To find the book with highest cost and display the details of that book\n4.Total Price\n\n5.Search book\n6.Exit\nEnter your choice:"))
    if ch == 1:
        for i in range(3):
            print("\nEnter the details of book", i+1, "\n")
            accno = int(input("Enter the account number:"))
            callno = int(input("Enter the call number:"))
            title = input("Enter the title of the book:")
            author = input("Enter the author of the book:")
            price = int(input("Enter the price of the book:"))
            lib.add(accno, callno, title, author, price)
    if ch == 2:
        lib.disp()
    elif ch == 3:
        lib.max()
    elif ch == 4:
        lib.total()
    elif ch == 5:
        book=input("ENter title")
        lib.srch(book)
    elif ch ==6:
        cursor.close()
        break
    else:
        print("\nInvalid choice")
```

```python
# 5.  Write a GUI program using frame to accept comma seperated list of numbers using a text box and display the maximum among them in a text box (use palce layout)

from tkinter import *


def findsum(num1):
    global number2
    n1 = map(int, num1.split(","))
    max1 = max(n1)
    number2.set(str(max1))
    return


root = Tk()
root.title("Finding maximum")
root.geometry("700x500")


Label(root, text="Enter comma seperated numbers").place(x=20, y=60)
e1 = Entry(root)
e1.place(x=200, y=60)

Label(root, text="Maximum number is").place(x=20, y=200)
Entry(root, textvariable=number2).place(x=200, y=200)

b1 = Button(root, text="Find Max", command=lambda: findsum(e1.get()))
b1.place(x=200, y=300)

root.mainloop()
```
