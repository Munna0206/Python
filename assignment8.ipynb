```python
# 1.Create a class called Employee with member data empno, name, depname, designation, age and salary and perform the following function
# 1.To accept details of 5 employees
# 2.To display details of all the employees
# 3.To search for an employee among all employees and display the details of that employee
# 4.To find the total salary of all the employee.

class Employee():
    def __init__(self, empno, ename, depname, designation, age, salary):
        self.empno = empno
        self.ename = ename
        self.depname = depname
        self.designation = designation
        self.age = age
        self.salary = salary

    def printEmp(self):
        print("\nEmployee number:", self.empno)
        print("Employee name:", self.ename)
        print("Employee department name:", self.depname)
        print("Employee designation:", self.designation)
        print("Employee age:", self.age)
        print("Employee salary:", self.salary)

emp = []
total=0
for i in range(3):
    print("\nEnter the details of employee", i+1, "\n")
    empno = int(input("Enter the employee number:"))
    ename = input("Enter the employee name:")
    depname = input("Enter the employee department:")
    designation = input("Enter the employee designation:")
    age = int(input("Enter the employee age:"))
    salary = int(input("Enter the employee salary:"))
    total=total+salary
    emp.append(Employee(empno, ename, depname, designation, age, salary))

while True:
    ch = int(input("\n1.Display all the employees\n2.search for an employee\n3.find total salary of all employee\n4.exit\nEnter your choice:"))
    if ch == 1:
        for i in range(3):
            emp[i].printEmp()
    elif ch == 2:
        name = input("Enter the employee name to search:")
        index = None
        for i in range(3):
            if name == emp[i].ename:
                index = i
                emp[index].printEmp()
                break
        if index != None:
            print("Employee not found")
    elif ch == 3:
        print("Total salary of all employee is:", total)
    elif ch == 4:
        break
    else:
        print("Invalid choice")
```

```python

# 3.  Write a GUI program to design a simple calculator (use the grid layout)

# 3.  Write a GUI program to design a simple calculator (use the grid layout)
# importing all the required modules from tkinter
from tkinter import Tk, StringVar, Frame, Entry, Button, TOP, RIGHT

win = Tk()
win.geometry("312x324")
win.resizable(0, 0)
win.title("Calculator")

expression = ""


def btn_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)


def bt_clear():
    global expression
    expression = ""
    input_text.set("")


def bt_equal():
    global expression
    result = str(eval(expression))
    input_text.set(result)
    expression = ""


def createButton(val, r, c, ch):
    Button(btns_frame, text=val, fg="black", width=10, height=3, command=lambda: btn_click(ch)).grid(row=r, column=c, padx=1, pady=1)


input_text = StringVar()

input_frame = Frame(win, width=312, height=50)

input_frame.pack(side=TOP)

input_field = Entry(input_frame,textvariable=input_text, width=50)

input_field.grid(row=0, column=0)

input_field.pack(ipady=10)

btns_frame = Frame(win, width=312, height=272.5, bg="grey")

btns_frame.pack()

# first row

Button(btns_frame, text="C", width=43, height=3, cursor="hand2",
       command=lambda: bt_clear()).grid(row=0, column=0, columnspan=4, padx=1, pady=1)

# second row
createButton("7", 1, 0, 7)
createButton("8", 1, 1, 8)
createButton("9", 1, 2, 9)
createButton("*", 1, 3, "*")
# third row
createButton("4", 2, 0, 4)
createButton("5", 2, 1, 5)
createButton("6", 2, 2, 6)
createButton("-", 2, 3, "-")
# fourth row
createButton("1", 3, 0, 1)
createButton("2", 3, 1, 2)
createButton("3", 3, 2, 3)
createButton("+", 3, 3, "+")
# fourth row
createButton("0", 4, 0, 0)
createButton(".", 4, 1, ".")
Button(btns_frame, text="=", width=10, height=3,command=lambda: bt_equal()).grid(row=4, column=2, padx=1, pady=1)
createButton("/", 4, 3, "/")

win.mainloop()
```

```python
# 5.  Write a GUI program using frame to accept comma seperated list of numbers using a text box and display the maximum among them in a text box (use palce layout)

from tkinter import *

def findsum(num1):
    global number2
    n1 = map(int, num1.split(","))
    max1 = max(n1)
    number2.set(str(max1))
    return


root = Tk()
root.title("Finding maximum")
root.geometry("700x500")
root.configure(bg="red")

number2=StringVar()
Label(root, text="Enter comma seperated numbers").place(x=20, y=60)
e1 = Entry(root)
e1.place(x=200, y=60)

Label(root, text="Maximum number is").place(x=20, y=200)
Entry(root, textvariable=number2).place(x=200, y=200)

b1 = Button(root, text="Find Max", command=lambda: findsum(e1.get()))
b1.place(x=200, y=300)

root.mainloop()
```

```python
# 6. Write a GUI program using menu to perform the following operations on Employee table
# i. Insert employee record
# ii. Update specified record
# iii. Delete Record
# iv. Display the records

# The menu items under menu ‘Operations’ are 1. Insert, 2. Update, 3. Delete, 4. Display, 5.Quit
import sqlite3
from tkinter import *

cursor=sqlite3.connect("newdata.db")
cursor.execute("create table if not exists emp(empno integer not null,ename text not null,salary integer not null)")
window=Tk()
empno=StringVar()
empname=StringVar()
salary=StringVar()
data=StringVar()

def add():
    cursor.execute("insert into emp values(?,?,?)",(int(empno.get()),empname.get(),int(salary.get())))
    cursor.commit()
    disp()
    clear()

def clear():
    empno.set("")
    empname.set("")
    salary.set("")

def disp():
    t1.delete('1.0',END)
    row=cursor.execute("select * from emp")
    for i in row:
        t1.insert(END,i)
        t1.insert(END,"\n")
    clear()

def update():
    cursor.execute("update emp set ename=?,salary=? where empno=?",(empname.get(),int(salary.get()),int(empno.get())))
    disp()
    clear()

def delete():
    cursor.execute("delete from emp where empno=?",(int(empno.get()),))
    disp()
    clear()

def exit():
    cursor.close()

Label(window,text="Empno").grid(row=0,column=0)
Label(window,text="Empname").grid(row=1,column=0)
Label(window,text="Salary").grid(row=2,column=0)
Entry(window,textvariable=empno).grid(row=0,column=1)
Entry(window,textvariable=empname).grid(row=1,column=1)
Entry(window,textvariable=salary).grid(row=2,column=1)
t1=Text(window)
t1.grid(row=3,column=0,columnspan=2)
Button(window,text='Add',command=add).grid(row=4,column=1)
Button(window,text='exit',command=exit).grid(row=4,column=0)
Button(window,text='Delete',command=delete).grid(row=5,column=1)
Button(window,text='Update',command=update).grid(row=6,column=1)
mainloop()
```
