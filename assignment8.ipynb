```python
# 1.Create a class called Employee with member data empno, name, depname, designation, age and salary and perform the following function
# 1.To accept details of 5 employees
# 2.To display details of all the employees
# 3.To search for an employee among all employees and display the details of that employee
# 4.To find the total salary of all the employee.

class Employee():
    def __init__(self, empno, ename, depname, designation, age, salary):
        self.empno = empno
        self.ename = ename
        self.depname = depname
        self.designation = designation
        self.age = age
        self.salary = salary

    def printEmp(self):
        print("\nEmployee number:", self.empno)
        print("Employee name:", self.ename)
        print("Employee department name:", self.depname)
        print("Employee designation:", self.designation)
        print("Employee age:", self.age)
        print("Employee salary:", self.salary)

    def getName(self):
        return self.ename

    def getSal(self):
        return self.salary


emp = []

for i in range(3):
    print("\nEnter the details of employee", i+1, "\n")
    empno = int(input("Enter the employee number:"))
    ename = input("Enter the employee name:")
    depname = input("Enter the employee department:")
    designation = input("Enter the employee designation:")
    age = int(input("Enter the employee age:"))
    salary = int(input("Enter the employee salary:"))
    emp.append(Employee(empno, ename, depname, designation, age, salary))

while True:
    ch = int(input("\n1.Display all the employees\n2.search for an employee\n3.find total salary of all employee\n4.exit\nEnter your choice:"))
    if ch == 1:
        for i in range(3):
            emp[i].printEmp()
    elif ch == 2:
        name = input("Enter the employee name to search:")
        index = None
        for i in range(3):
            if name == emp[i].getName():
                index = i
                break
        if index != None:
            emp[index].printEmp()
        else:
            print("Employee not found")
    elif ch == 3:
        totsal = 0
        for i in range(3):
            totsal = totsal+emp[i].getSal()
        print("Total salary of all employee is:", totsal)
    elif ch == 4:
        break
    else:
        print("Invalid choice")
```

```python
# 2.  Create a class called Library with Accno, Callno, Title of the book, author, price of the book. Perform the following
#      1.	To accept the details of 5 books and store it in database.
#      2.	To display the details of all books.
#      3.	To find the book with highest cost and display the details of that book.
#      4.   To display the total cost of all books put together.
#      5.   To search for a perticular book


class Library():
    def __init__(self, accno, callno, title, author, price):
        self.accno = accno
        self.callno = callno
        self.title = title
        self.author = author
        self.price = price

    def printBook(self):
        print("\nAccount number:", self.accno)
        print("Call number:", self.callno)
        print("Title of the book:", self.title)
        print("Author of the book:", self.author)
        print("Price of the book:", self.price)

    def getBook(self):
        return self.title

    def getPrice(self):
        return self.price


libr = []

for i in range(3):
    print("\nEnter the details of book", i+1, "\n")
    accno = int(input("Enter the account number:"))
    callno = int(input("Enter the call number:"))
    title = input("Enter the title of the book:")
    author = input("Enter the author of the book:")
    price = int(input("Enter the price of the book:"))
    libr.append(Library(accno, callno, title, author, price))

while True:
    ch = int(input("\n1.Display details of all the books\n2.Find book with highest cost\n3.find total price of all books\n4.Search for a particular book\n5.Exit \nEnter your choice:"))
    if ch == 1:
        for i in range(3):
            libr[i].printBook()
    elif ch == 2:
        max1 = 0
        index = None
        for i in range(3):
            if libr[i].getPrice() > max1:
                max1 = libr[i].getPrice()
                index = i
        print("\nThe book with maximum cost is:")
        libr[index].printBook()
    elif ch == 3:
        totcost = 0
        for i in range(3):
            totcost = totcost+libr[i].getPrice()
        print("\nTotal cost of all book is:", totcost)
    elif ch == 4:
        name = input("\nEnter the book title to search:")
        index = None
        for i in range(3):
            if name == libr[i].getBook():
                index = i
                break
        if index != None:
            libr[index].printBook()
        else:
            print("\nBook not found")
    elif ch == 5:
        break
    else:
        print("\nInvalid choice")
```

```python

# 3.  Write a GUI program to design a simple calculator (use the grid layout)
# importing all the required modules from tkinter
from tkinter import Tk, StringVar, Frame, Entry, Button, TOP, RIGHT

win = Tk()
win.geometry("312x324")
win.resizable(0, 0)
win.title("Calculator")

expression = ""


def btn_click(item):
    global expression
    expression = expression + str(item)
    input_text.set(expression)


def bt_clear():
    global expression
    expression = ""
    input_text.set("")


def bt_equal():
    global expression
    result = str(eval(expression))
    input_text.set(result)
    expression = ""


def createButton(val, r, c, ch):
    Button(btns_frame, text=val, fg="black", width=10, height=3, bd=0, bg="#fff",
           cursor="hand2", command=lambda: btn_click(ch)).grid(row=r, column=c, padx=1, pady=1)


input_text = StringVar()

input_frame = Frame(win, width=312, height=50, bd=0,
                    highlightbackground="black", highlightcolor="black", highlightthickness=2)

input_frame.pack(side=TOP)

input_field = Entry(input_frame, font=('arial', 18, 'bold'),
                    textvariable=input_text, width=50, bg="#eee", bd=0, justify=RIGHT)

input_field.grid(row=0, column=0)

input_field.pack(ipady=10)

btns_frame = Frame(win, width=312, height=272.5, bg="grey")

btns_frame.pack()

# first row

Button(btns_frame, text="C", width=43, height=3, fg="black", bd=0, bg="#fff", cursor="hand2",
       command=lambda: bt_clear()).grid(row=0, column=0, columnspan=4, padx=1, pady=1)

# second row
createButton("7", 1, 0, 7)
createButton("8", 1, 1, 8)
createButton("9", 1, 2, 9)
createButton("*", 1, 3, "*")
# third row
createButton("4", 2, 0, 4)
createButton("5", 2, 1, 5)
createButton("6", 2, 2, 6)
createButton("-", 2, 3, "-")
# fourth row
createButton("1", 3, 0, 1)
createButton("2", 3, 1, 2)
createButton("3", 3, 2, 3)
createButton("+", 3, 3, "+")
# fourth row
createButton("0", 4, 0, 0)
createButton(".", 4, 1, ".")
Button(btns_frame, text="=", fg="black", width=10, height=3, bd=0, bg="#eee",
       cursor="hand2", command=lambda: bt_equal()).grid(row=4, column=2, padx=1, pady=1)
createButton("/", 4, 3, "/")

win.mainloop()
```

```python
# 4.  Write Program using Tkinder to accept the color from user and change the background color according to the accepted color. (use the pack layout)
from tkinter import Tk, Label, Entry, TOP, Button, Frame, font

win = Tk()
win.geometry("400x200")
win.configure(bg='blue')
win.title("Change Background")


def changebg(color):
    win.configure(bg=color)


myFont = font.Font(size=8)
frame = Frame(win, borderwidth=3)
frame.pack(padx=10, pady=10)
e1 = Entry(frame, bd=2)
e1.pack(side=TOP)
B1 = Button(win, text="Change", fg="black", width=10, height=1, bd=0, bg="#fff",
            cursor="hand2", command=lambda: changebg(e1.get()))
B1.pack(side=TOP)
B1['font'] = myFont
win.mainloop()
```

```python
# 5.  Write a GUI program using frame to accept comma seperated list of numbers using a text box and display the maximum among them in a text box (use palce layout)

from tkinter import Entry, Tk, StringVar, Label, Button


def findsum(num1):
    global number2
    n1 = map(int, num1.split(","))
    max1 = max(n1)
    number2.set(str(max1))
    return


root = Tk()
root.title("Finding maximum")
root.geometry("700x500")

number2 = StringVar()

Label(root, text="Enter comma seperated numbers").place(x=20, y=60)
e1 = Entry(root)
e1.place(x=200, y=60)

Label(root, text="Maximum number is").place(x=20, y=200)
Entry(root, textvariable=number2).place(x=200, y=200)


b1 = Button(root, text="Find Max", command=lambda: findsum(e1.get()))
b1.place(x=200, y=300)

root.mainloop()
```

```python
from tkinter import *
from tkinter.ttk import *
import tkinter.messagebox
import sqlite3

root=Tk()
root.geometry('1200x700')
root.title('Employee database')
root.config(bg='#223467')


def database():
    conn=sqlite3.connect("empdata.db")
    cursor=conn.cursor()
    cursor.execute("create table if not exists `emp`(eno text not null primary key,ename text not null,dept text not null,age integer not null,desig text not null,salary real not null)")
    cursor.close()
    conn.close()

def indata():
    if eno.get()=="" or ename.get()=="" or depname.get()=="" or age.get()<=0 or desig.get()=="" or salary.get()<=0:
        res=tkinter.messagebox.showinfo("msg","please enter the field")
    else:
        conn=sqlite3.connect("empdata.db")
        cursor=conn.cursor()
        cursor.execute("insert into emp values(?,?,?,?,?,?)",(str(eno.get()),str(ename.get()),str(depname.get()),int(age.get()),str(desig.get()),float(salary.get())))
        res=tkinter.messagebox.showinfo("msg","record inserted")
        conn.commit()
        cursor.close()
        conn.close()
    
def updata():
      if eno.get()=="" or ename.get()=="" or depname.get()=="" or age.get()<=0 or desig.get()=="" or salary.get()<=0:
          res=tkinter.messagebox.showinfo("msg","please enter the field")
      else:
          conn=sqlite3.connect("empdata.db")
          cursor=conn.cursor()
          cursor.execute("update emp set ename=?, dept=?, age=?, desig=?, salary=? where eno=?",(str(ename.get()),str(depname.get()),int(age.get()),str(desig.get()),float(salary.get()),str(eno.get())))
          
          res=tkinter.messagebox.showinfo("msg","record updated")
          conn.commit()
          cursor.close()
          conn.close()
          
          
def deldata():
      if eno.get()=="" or ename.get()=="" or depname.get()=="" or age.get()<=0 or desig.get()=="" or salary.get()<=0:
          res=tkinter.messagebox.showinfo("msg","please enter the field")
      else:
          conn=sqlite3.connect("empdata.db")
          cursor=conn.cursor()
          cursor.execute("delete from  emp where eno=?",(str(ename.get()),str(depname.get()),int(age.get()),str(desig.get()),float(salary.get()),str(eno.get())))
          
          res=tkinter.messagebox.showinfo("msg","record updated")
          conn.commit()
          cursor.close()
          conn.close()
 
        
def viewdata():
    if eno.get()=="":
        res=tkinter.messagebox.showinfo("msg","please enter the field")
    else:
        conn=sqlite3.connect("empdata.db")
        cursor=conn.cursor()
        cursor.execute("select * from emp where eno="+eno.get())
        rows=cursor.fetchall()
        ta.delete("1.0", END)
        for row in rows:
            ta.insert(END, "\n")
            ta.insert(END, row)
        conn.commit()
        cursor.close()
        conn.close()
                
def viewalldata():
        conn=sqlite3.connect("empdata.db")
        cursor=conn.cursor()
        cursor.execute("select * from emp")
        rows=cursor.fetchall()
        ta.delete("1.0", END)
        for row in rows:
            ta.insert(END, "\n")
            ta.insert(END, row)
       
        conn.commit()
        cursor.close()
        conn.close()

def tot():
    t=0
    conn=sqlite3.connect("empdata.db")
    cursor=conn.cursor()
    cursor.execute("select * from emp")
    rows=cursor.fetchall()
    ta.delete("1.0", END)
    for row in rows:
        t=t+int(row[5])
    ta.insert(END,"total="+str(t))
    conn.commit()
    cursor.close()
    conn.close()

def highsal():
    hs=0
    conn=sqlite3.connect("empdata.db")
    cursor=conn.cursor()
    cursor.execute("select * from emp")
    rows=cursor.fetchall()
    ta.delete("1.0", END)
    hs=rows[0]
    for row in rows:
       if row[5]>hs[5]:
           hs=row
    
    ta.insert(END,"total="+str(hs))
    conn.commit()
    cursor.close()
    conn.close()

    
    

database()

eno=StringVar()
ename=StringVar()
depname=StringVar()
desig=StringVar()
age=IntVar()
salary=DoubleVar()

l1=Label(root,text="EMPNO",font=('arial',12))
l1.place(x=50,y=50)

t1=Entry(root,textvariable=eno)
t1.place(x=160,y=50)

l2=Label(root,text="ENAME",font=('arial',12))
l2.place(x=50,y=100)

t2=Entry(root,textvariable=ename)
t2.place(x=160,y=100)

l3=Label(root,text='DEPT',font=('arial',12))
l3.place(x=50,y=150)

t3=Entry(root,textvariable=depname)
t3.place(x=160,y=150)

l4=Label(root,text='AGE',font=('arial',12))
l4.place(x=50,y=200)

t4=Entry(root,textvariable=age)
t4.place(x=160,y=200)


l5=Label(root,text='DESIG',font=('arial',12))
l5.place(x=50,y=250)
combo=Combobox(root,textvariable=desig)
combo['value']=("Manager","clerk","staff")
combo.place(x=160,y=250)

l6=Label(root,text='SALARY',font=('arial',12))
l6.place(x=50,y=300)
t6=Entry(root,textvariable=salary)
t6.place(x=160,y=300)

bt=Button(root,text="INSERT",command=indata)
bt.place(x=100,y= 400)

bt1=Button(root,text="view",command=viewdata)
bt1.place(x=100,y= 500)

bt2=Button(root,text="view all",command=viewalldata)
bt2.place(x=200,y= 400)

bt3=Button(root,text="tot",command=tot)
bt3.place(x=200,y= 500)

bt4=Button(root,text="update",command=updata)
bt4.place(x=300,y=500)

bt5=Button(root,text="highsal",command=highsal)
bt5.place(x=500,y=500)

bt6=Button(root,text="delete",command=deldata)
bt6.place(x=500,y=500)
#textarea
ta=Text(root)
ta.place(x=400,y=50)

root.mainloop()
```
