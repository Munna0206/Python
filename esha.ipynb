```python
import random
attempt=0
while (not(attempt in range(1,11))):
     attempt=int(input("How many attempts do you want? [1-10]"))

print("Selecting words:")
fp=open("words.txt","r")
lst=fp.read().split()
randm=random.randint(0,len(lst))
word=lst[randm].lower()
fp.close()
prev=set()
incorrect=set()
unguess=""
for i in word:
    unguess=unguess+"*"
while(attempt!=0):
    print("\nRemaning attempt:",attempt)
    print("Word:",unguess)
    print("Previous Correct Guess:",prev)
    print("Previous Incorrect Guess:",incorrect)
    guess=input("Guess a character")
    if (guess in prev):
        print(guess,"has already guessed")
    elif (guess in word):
        prev.add(guess)
        unguess=""
        for i in word:
            if(i in prev):
                unguess=unguess+i
            else:
                unguess=unguess+"*"
        if(not('*' in unguess)):
            print("\nYou won\nWord is:",word)
            break
    else:
        print("Invalid character")
        incorrect.add(guess)
        attempt-=1
else:
    print("\nYou Lost\nActual Word is:",word)
```
